```r
sampled_address = address %>%
    sample_n(size=50)
sampled_address
```

## Testing Regex

#### 1.
Pattern like \[number\] \[name\] \[street\]

```r
# Match
sampled_address %>%
    filter(grepl("[0-9]+ [A-Z]+ ST", address))
```

```r
# Not match
sampled_address %>%
    filter(!grepl("[0-9]+ [A-Z]+ ST", address))
```

#### 2.
Pattern like \[number\] th street

```r
# Match
sampled_address %>%
    filter(grepl("[0-9]+ [A-Z]+ ST", address))
```

```r
# Not match
sampled_address %>%
    filter(!grepl("[0-9]+ [A-Z]+ ST", address))
```

```r
pattern <- "(ST|RD|ROAD)"
sampled_address %>% 
    filter(!grepl(pattern, address),  # Not contain ST, RD or ROAD
           !grepl(" APT|UNIT [[:digit:]]+$", address)) 
            # Not contain APT or UNIT followed by a space, numbers in the end of the string.
```
